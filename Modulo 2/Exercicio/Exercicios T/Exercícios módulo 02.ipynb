1. Crie uma classe Bola cujos atributos são cor e raio. Crie um método que imprime a cor da bola. Crie um método para calcular a área dessa bola. Crie um método para calcular o volume da bola. Crie um objeto dessa classe e calcule a área e o volume, imprimindo ambos em seguida. 
    Obs.:

    Área da esfera = 4*3.14*r*r;

    Volume da esfera = 4*3.14*r*r*r/3

```python
class Bola:
    def __init__(self, cor, raio):
        '''
        Cria uma representação da bola

        Parâmetros:
        -----------
        cor: str
            Cor atribuída à bola
        raio: float
            Valor do raio
        '''
        self.cor = cor
        self.raio = raio
    
    def imprime_cor(self):
        '''
        Imprime a cor da bola
        '''
        print('A bola é', self.cor)
    
    def calcula_area(self):
        '''
        Calcula a área da bola
        '''
        return 4*3.14*self.raio**2
    
    def calcula_volume(self):
        '''
        Calcula o volume da bola
        '''
        return 4*3.14*self.raio**3/3
    
#Exemplo de uso:
bola1 = Bola('azul', 2)
bola1.imprime_cor()
print('A área da bola é', bola1.calcula_area())
print('O volume da bola é', bola1.calcula_volume())
```

3. Crie uma classe Funcionario cujos atributos são nome e e-mail. Guarde as horas trabalhadas em um dicionário cujas chaves são o mês em questão e, em outro dicionário, guarde o salário por hora relativo ao mês em questão. Crie um método que retorna o salário mensal do funcionário.

```python
class Funcionario:
    
    def __init__(self, nome, email):
        '''
        Cria uma representação do funcionário

        Parâmetros:
        -----------
        nome: str
            Nome do funcionário
        email: str
            Endereço de e-mail
        '''
        self.nome = nome
        self.email = email
        
        self.horas_trabalhadas = {}
        self.salario_hora = {}
        
    def atualiza_horas_trabalhadas(self, data, horas):
        '''
        Armazena a quantidade de horas trabalhadas em um determinado dia

        Parâmetros:
        -----------
        data: str
            Data
        horas: float
            Quantidade de horas trabalhadas
        '''
        self.horas_trabalhadas[data] = horas
        
    def atualiza_salario_hora(self, data, valor):
        '''
        Armazena o valor da hora trabalhada em um determinado dia

        Parâmetros:
        -----------
        data: str
            Data
        valor: float
            Valor da hora trabalhada
        '''
        self.salario_hora[data] = valor
        
    def calcula_salario(self, data):
        '''
        Calcula o valor do salário com base na data informada

        Parâmetros:
        -----------
        data: str
            Data
        '''
        return self.horas_trabalhadas[data] * self.salario_hora[data]

#Exemplo de uso
pedro = Funcionario('Pedro', 'pedro@123.com')
pedro.atualiza_horas_trabalhadas('13/11/21', 10)
pedro.atualiza_salario_hora('13/11/21', 15.00)
pedro.calcula_salario('13/11/21')
```

4. Crie uma classe Televisor cujos atributos são:

    a. fabricante;

    b. modelo;

    c. canal atual;

    d. lista de canais; e

    e. volume.

    Faça métodos para aumentar/diminuir volume, trocar o canal e sintonizar um novo canal, que adiciona um novo canal à lista de canais (somente se esse canal não estiver nessa lista). No atributo lista de canais, devem estar armazenados todos os canais já sintonizados dessa TV.

    Obs.: O volume não pode ser menor que zero e maior que cem; só se pode trocar para um canal que já esteja na lista de canais.

```python
class Televisor:
        
    def __init__(self, fabricante, modelo):
        '''
        Cria uma representação da TV

        Parâmetros:
        -----------
        fabricante: str
            Nome do fabricante
        modelo: str
            Modelo da TV
        canal_atual: int
            Número do canal atual
        lista_de_canais: list
            Lista contendo os canais já sintonizados
        volume: int
            Volume atual
        '''
        self.fabricante = fabricante
        self.modelo = modelo
        self.canal_atual = ''
        self.lista_de_canais = []
        self.volume = 0
    
    def aumenta_volume(self, valor):
        '''
        Aumenta o volume com o valor indicado, respeitando o limite máximo de 100
       
        Parâmetros:
        -----------
        valor: int
            Valor para o volume
        '''
        if self.volume + valor <= 100:
            self.volume += valor
        else:
            self.volume = 100
            
    def diminui_volume(self, valor):
        '''
        Diminui o volume com o valor indicado, respeitando o limite mínimo de 0
       
        Parâmetros:
        -----------
        valor: int
            Valor para o volume
        '''
        if self.volume - valor >= 0:
            self.volume -= valor
        else:
            self.volume = 0
            
    def troca_canal(self, canal):
        '''
        Altera para o canal indicado, se o mesmo estiver na lista de canais
       
        Parâmetros:
        -----------
        canal: str
            Nome do canal
        '''
        if canal.lower() in self.lista_de_canais:
            self.canal_atual = canal.lower()
        else:
            self.canal_atual = self.canal_atual
            
    def sintoniza_canal(self, canal):
        '''
        Adiciona o canal indicado à lista de canais, observando
        se o mesmo já não se encontra na lista de canais
       
        Parâmetros:
        -----------
        canal: str
            Nome do canal
        '''
        if canal.lower() not in self.lista_de_canais:
            self.lista_de_canais.append(canal.lower())  
            
#Exemplo de uso:
tv1 = Televisor('LG', 'teste')
tv1.sintoniza_canal('SBT')
tv1.sintoniza_canal('Globo')
tv1.aumenta_volume(90)
print(tv1.volume)
tv1.diminui_volume(110)
print(tv1.volume)
tv1.troca_canal('GlOBo')
tv1.troca_canal('AXN')
tv1.canal_atual
```

6. O módulo time possui a função time.sleep(x), que faz seu programa “dormir” por x segundos. Utilizando essa função, crie uma classe Cronômetro e faça um programa que cronometre o tempo.

```python
import time

class Cronometro:
        
    def __init__(self, segundos = 0, minutos = 0, horas = 0):
        '''
        Cria uma representação do cronômetro

        Parâmetros:
        -----------
        segundos: int
            Valor referente aos segundos
        minutos: int
            Valor referente aos minutos
        horas: int
            Valor referente às horas
        '''
        self.segundos = segundos
        self.minutos = minutos
        self.horas = horas
        
    def __repr__(self):
        '''
        Formata a apresentação do cronômetro
        '''
        return "%.2d:%.2d:%.2d" %(self.horas, self.minutos, self.segundos)
    
    def incremento(self):
        '''
        Realiza a contagem (incremento) dos segundos/minutos/horas
        '''
        if self.segundos >= 60:
            self.segundos = 0
            self.minutos += 1
        if self.minutos >= 60:
            self.minutos = 0
            self.horas += 1
        self.segundos += 1
        
    def run(self):
        '''
        Inicia o cronômetro
        '''
        while True:  
            print(self)
            self.incremento()
            time.sleep(1)
            
#Exemplo de uso:
cronometro1 = Cronometro()
cronometro1.run()
```

7. Crie uma modelagem de classes para uma agenda capaz de armazenar contatos. Através dessa agenda é possível incluir, remover, buscar e listar contatos já cadastrados.

```python
class Contato:
    
    def __init__(self, nome, telefone):
        '''
        Cria uma representação do contato

        Parâmetros:
        -----------
        nome: str
            Nome da pessoa
        telefone: str
            Telefone
        '''
        self.nome = nome
        self.telefone = telefone
    
    def __str__(self):
        return f'Nome: {self.nome}\nTelefone: {self.telefone}\n'
        
class Agenda:

    def __init__(self):
        '''
        Cria uma representação da agenda

        Parâmetros:
        -----------
        contatos: dict
            Dicionário contendo as informações dos contatos
        '''
        self.contatos = {}
        
    def incluir_contato(self, novo_contato):
        '''
        Adiciona um novo contato na agenda

        Parâmetros:
        -----------
        novo_contato: object
            Objecto da classe Contato
        '''
        self.contatos[novo_contato.nome] = novo_contato
        
    def remover_contato(self, contato):
        '''
        Remove um contato na agenda

        Parâmetros:
        -----------
        contato: object
            Objecto da classe Contato a ser removido
        '''
        self.contatos.pop(contato.nome)
    
    def listar_contatos(self):
        '''
        Exibe as informações de todos os contatos cadastrados
        '''
        for contato in self.contatos.values():
            print(contato)

    def buscar_contato(self, nome):
        '''
        Exibe as informações do contato selecionado
        
        Parâmetros:
        -----------
        contato: str
            Nome do contato
        '''
        for contato in self.contatos.values():
            if contato.nome == nome:
                print(contato)
                
#Exemplo de uso:
thiago = Contato('Thiago', '123456')
thiagopastre = Contato('Thiago Pastre', '000')
lucas = Contato('Lucas', '789')

agenda1 = Agenda()
agenda1.incluir_contato(thiago)
agenda1.incluir_contato(thiagopastre)
agenda1.incluir_contato(lucas)
agenda1.listar_contatos()
agenda1.remover_contato(lucas)
agenda1.buscar_contato('Thiago Pastre')
```

9. Com base no exercício anterior, crie um sistema de cadastro e a classe Cliente. Seu programa deve perguntar se o usuário quer cadastrar um novo cliente, alterar um cadastro ou sair.

Dica: Você pode fazer esse exercício criando uma classe Sistema, que irá controlar o sistema de cadastros. Essa classe deve ter o atributo cadastro e os métodos para imprimir os cadastrados, cadastrar um novo cliente, alterar um cadastro ou sair.

```python
class Cliente:
    
    def __init__(self, nome, idade, email):
        '''
        Cria o cliente
        
        Parâmetros:
        -----------
        nome: str
            Nome do cliente
        idade: int
            Idade
        email: str
            E-mail
        '''
        self.nome = nome
        self.idade = idade
        self.email = email
    
    def __str__(self):
        return f'Nome: {self.nome}\nIdade: {self.idade}\nE-mail: {self.email}\n'
    
class Sistema:
    
    def __init__(self):
        '''
        Controla o sistema de cadastros
        '''
        self.cadastro = {}
        
    def cadastrar_novo_cliente(self, novo_cliente):
        '''
        Adiciona um novo cliente

        Parâmetros:
        -----------
        novo_cliente: object
            Objecto da classe Cliente
        '''
        self.cadastro[novo_cliente.nome.lower()] = novo_cliente
        
    def imprimir_clientes_cadastrados(self):
        '''
        Exibe as informações de todos os clientes cadastrados
        '''
        for cliente in self.cadastro.values():
            print(cliente)
            
    def alterar_cadastro_cliente(self, cliente, alteracao, novo_valor):
        '''
        Altera o cadastro do cliente

        Parâmetros:
        -----------
        cliente: object
            Objecto da classe Cliente
        alteracao = str
            Refere-se à qual informação será alterada
        novo_valor = str para nome e email / int para idade
        '''
        if alteracao == 'nome':
            self.cadastro[cliente].nome = novo_valor
        
        elif alteracao == 'idade':
            self.cadastro[cliente].idade = novo_valor
            
        elif alteracao == 'email':
             self.cadastro[cliente].email = novo_valor
                
    def run(self):
        opcao = 10
        while opcao != 0:
            print('\n1. Cadastrar um novo cliente',
                  '\n2. Alterar um cadastro',
                  '\n3. Listar clientes',
                  '\n0. Sair')
            opcao = int(input('Selecione a opção desejada: '))
            
            if opcao == 1:
                nome = input('Nome: ')
                idade = input('Idade: ')
                while not idade.isdigit():
                    print('Valor inválido, digite novamente')
                    idade = input('Idade: ')
                email = input('E-mail: ')
                novo = Cliente(nome.title(), idade, email.lower())
                self.cadastrar_novo_cliente(novo)
                print('Cliente cadastrado com sucesso!\n')
                self.run()
                
            if opcao == 2:
                cliente = input('Informe o nome do cliente: ')
                if cliente.lower() in self.cadastro:
                    alteracao = input('O que deseja alterar? (nome, idade, email): ')
                    if alteracao.lower() == 'nome':
                        novo_nome = input('Informe o novo nome: ')
                        self.alterar_cadastro_cliente(cliente.lower(), alteracao, novo_nome.title())
                        print('Alteração realizada com sucesso')
                        self.run()
                    
                    elif alteracao.lower()  == 'idade':
                        nova_idade = input('Informe a nova idade: ')
                        while not nova_idade.isdigit():
                            print('Valor inválido, digite novamente')
                            nova_idade = input('Idade: ')
                        self.alterar_cadastro_cliente(cliente.lower(), alteracao, nova_idade)
                        print('Alteração realizada com sucesso')
                        self.run()
                        
                    elif alteracao.lower() == 'email':
                        novo_email = input('Informe o novo e-mail: ')
                        self.alterar_cadastro_cliente(cliente.lower(), alteracao, novo_email.lower())
                        print('Alteração realizada com sucesso')
                        self.run()
                        
                    else:
                        print('Opção inválida')
                        self.run()
                else:
                    print('Cliente não cadastrado')
                    self.run()
                
            if opcao == 3:
                if len(self.cadastro) == 0:
                    print('Cadastro vazio\n')
                self.imprimir_clientes_cadastrados()
                self.run()
                
#Exemplo de uso:
sistema1 = Sistema()
sistema1.run()
```

```python
class Cliente:
    
    def __init__(self, nome, cpf):
        '''
        Cria o cliente
        
        Parâmetros:
        -----------
        nome: str
            Nome do cliente
        cpf: str
            CPF
        '''    
        self.nome = nome
        self.cpf = cpf
        
class ContaCorrente:
    
    def __init__(self, cliente):
        '''
        Cria a representação da conta corrente
        
        Parametros:
        -----------
        cliente = object
            Objeto da classe Cliente
        '''
        self.cliente = cliente
        self.saldo = 0
        
    def saque(self, valor):
        '''
        Realiza a dedução do valor na conta corrente, caso haja saldo suficiente
        
        Parametros:
        -----------
        valor = float
            Valor do saque
        '''
        if self.saldo - valor >= 0:
            self.saldo -= valor
        else:
            print('Saldo insuficiente.')
            
    def deposito(self, valor):
        '''
        Realiza a adição do valor na conta corrente
        
        Parametros:
        -----------
        valor = float
            Valor do depósito
        '''
        self.saldo += valor
        
    def transferencia(self, other, valor):
        '''
        Realiza a transferência do valor da conta corrente de self para other
        
        Parametros:
        -----------
        other = object
            Objeto da classe Cliente
        valor = float
            Valor da transferência
        '''
        if self.saldo - valor >= 0:
            self.saldo -= valor
            other.saldo += valor
        else:
            print('Saldo insuficiente.')
            
#Exemplo de uso:
thiago = Cliente('Thiago','9999999')
lucas = Cliente('Lucas', '888888')
conta_thiago = ContaCorrente(thiago)
conta_lucas = ContaCorrente(lucas)
conta_thiago.deposito(120)
print(conta_thiago.saldo)
print(conta_lucas.saldo)
conta_thiago.transferencia(conta_lucas, 120)
conta_thiago.saque(100)
print(conta_thiago.saldo)
print(conta_lucas.saldo)
```

11. Crie uma classe Fração cujos atributos são numerador (número de cima) e denominador (número de baixo).

Implemente os métodos de adição, subtração, multiplicação, divisão que retornam objetos do tipo Fração.

Implemente também o método _ repr _.

Implemente métodos para comparação: igualdade (==) e desigualdades (!=, <=, >=, < e >).

```python
class Fracao:

    def __init__(self, numerador, denominador):
        '''
        Cria a representação da fração
        
        Parametros:
        -----------
        numerador: float
            Numerador da fração
        denominador: float
            Denominador da fração
        '''
        if denominador == 0:
            raise ValueError("O denominador não pode ser 0!")
        self.numerador = numerador
        self.denominador = denominador
        
    def __repr__(self):
        return f'{self.numerador}/{self.denominador}'
    
    def __add__(self, other):
        num = self.numerador*other.denominador +\
        self.denominador*other.numerador
        den = self.denominador*other.denominador
        return Fracao(num, den)
    
    def __sub__(self, other):
        num = self.numerador*other.denominador -\
        self.denominador*other.numerador
        den = self.denominador*other.denominador 
        return Fracao(num, den)
    
    def __mul__(self, other):
        num = self.numerador*other.numerador
        den = self.denominador*other.denominador
        return Fracao(num, den)
    
    def __truediv__(self, other):
        num = self.numerador*other.denominador
        den = self.denominador*other.numerador
        return Fracao(num, den)
    
    def valor(self):
        return self.numerador/self.denominador
    
    def __eq__(self, other):
        return self.valor() == other.valor()
    
    def __lt__(self,other):
        return self.valor() < other.valor()
    
    def __le__(self,other):
        return self.valor() <= other.valor()
    
    def __gt__(self, other):
        return self.valor() > other.valor()
    
    def __ge__(self, other):
        return self.valor() >= other.valor()
    
#Exemplo de uso:
fracao1 = Fracao(1,16)
fracao2 = Fracao(7,9)
print(fracao1 + fracao2)
print(fracao1 - fracao2)
print(fracao1 * fracao2)
print(fracao1 / fracao2)
print(fracao1 > fracao2)
print(fracao1 >= fracao2)
print(fracao1 < fracao2)
print(fracao1 <= fracao2)
print(fracao1 == fracao2)
```

12. Crie uma classe Data cujos atributos são dia, mês e ano. Implemente métodos _ repr _ e para comparação: igualdade (==) e desigualdades (!=, <=, >=, < e >).

```python
class Data:
    
    def __init__(self, dia, mes, ano):
        self.dia = dia
        self.mes = mes
        self.ano = ano
        
    def __repr__(self):
        return "{}/{}/{}".format(
            self.dia,
            self.mes,
            self.ano
        )
    
    def __eq__(self, other):
        return self.dia == other.dia and \
                self.mes == other.mes and \
                self.ano == other.ano
    
    def __neq__(self, other):
        return not self == other
    
    def __lt__(self, other):
        if self.ano == other.ano:
            if self.mes == other.mes:
                if self.dia == other.dia:
                    return False
                else:
                    return self.dia < other.dia
            else:
                return self.mes < other.mes
        else:
            return self.ano < other.ano
        
    def __le__(self, other):
        return self < other or self == other
    
    def __gt__(self, other):
        return not self <= other
    
    def __ge__(self, other):
        return not self < other
    
#Exemplo de uso
data1 = Data(15,11,2021)
data2 = Data(22,10,1985)

print(data1 > data2)
print(data1 >= data2)
print(data1 < data2)
print(data1 <= data2)
print(data1 == data2)
print(data1 != data2)
```

```python
class Retangulo:
        
    def __init__(self, lado_a, lado_b):
        '''
        Cria uma representação do retângulo

        Parâmetros:
        -----------
        lado_a: float
            Valor da base do retângulo
        lado_b: float
            Valor da altura do retângulo
        '''
        self.lado_a = lado_a
        self.lado_b = lado_b
        
    def __repr__(self):
        return f'Lado A: {self.lado_a}\nLado B: {self.lado_b}\
                \nÁrea: {self.calcula_area()}'
    
    def calcula_area(self):
        '''
        Calcula a área do retângulo
        '''
        return self.lado_a * self.lado_b

#Exemplo de uso
retangulo1 = Retangulo(5, 2)
retangulo1
```

```python
class Funcionario:
    
    def __init__(self, nome, email):
        '''
        Cria uma representação do funcionário

        Parâmetros:
        -----------
        nome: str
            Nome do funcionário
        email: str
            Endereço de e-mail
        '''
        self.nome = nome
        self.email = email
        
        self.horas_trabalhadas = {}
        self.salario_hora = {}
        
    def __repr__(self):
        return f'Nome: {self.nome}\nEmail: {self.email}'
        
    def atualiza_horas_trabalhadas(self, data, horas):
        '''
        Armazena a quantidade de horas trabalhadas em um determinado dia

        Parâmetros:
        -----------
        data: str
            Data
        horas: float
            Quantidade de horas trabalhadas
        '''
        self.horas_trabalhadas[data] = horas
        
    def atualiza_salario_hora(self, data, valor):
        '''
        Armazena o valor da hora trabalhada em um determinado dia

        Parâmetros:
        -----------
        data: str
            Data
        valor: float
            Valor da hora trabalhada
        '''
        self.salario_hora[data] = valor
        
    def calcula_salario(self, data):
        '''
        Calcula o valor do salário com base na data informada

        Parâmetros:
        -----------
        data: str
            Data
        '''
        return self.horas_trabalhadas[data] * self.salario_hora[data]

#Exemplo de uso:
pedro = Funcionario('Pedro', 'pedro@123.com')
pedro
```

```python
class Televisor:
        
    def __init__(self, fabricante, modelo):
        '''
        Cria uma representação da TV

        Parâmetros:
        -----------
        fabricante: str
            Nome do fabricante
        modelo: str
            Modelo da TV
        canal_atual: int
            Número do canal atual
        lista_de_canais: list
            Lista contendo os canais já sintonizados
        volume: int
            Volume atual
        '''
        self.fabricante = fabricante
        self.modelo = modelo
        self.canal_atual = ''
        self.lista_de_canais = []
        self.volume = 0
        
    def __repr__(self):
        return f'Fabricante: {self.fabricante}\nModelo: {self.modelo}\
                \nCanal atual: {self.canal_atual}\nLista de canais sintonizados: {self.lista_de_canais}\
                \nVolume atual: {self.volume}'
    
    def aumenta_volume(self, valor):
        '''
        Aumenta o volume com o valor indicado, respeitando o limite máximo de 100
       
        Parâmetros:
        -----------
        valor: int
            Valor para o volume
        '''
        if self.volume + valor <= 100:
            self.volume += valor
        else:
            self.volume = 100
            
    def diminui_volume(self, valor):
        '''
        Diminui o volume com o valor indicado, respeitando o limite mínimo de 0
       
        Parâmetros:
        -----------
        valor: int
            Valor para o volume
        '''
        if self.volume - valor >= 0:
            self.volume -= valor
        else:
            self.volume = 0
            
    def troca_canal(self, canal):
        '''
        Altera para o canal indicado, se o mesmo estiver na lista de canais
       
        Parâmetros:
        -----------
        canal: str
            Nome do canal
        '''
        if canal.lower() in self.lista_de_canais:
            self.canal_atual = canal.lower()
        else:
            self.canal_atual = self.canal_atual
            
    def sintoniza_canal(self, canal):
        '''
        Adiciona o canal indicado à lista de canais, observando
        se o mesmo já não se encontra na lista de canais
       
        Parâmetros:
        -----------
        canal: str
            Nome do canal
        '''
        if canal.lower() not in self.lista_de_canais:
            self.lista_de_canais.append(canal.lower())  
            
#Exemplo de uso:
tv1 = Televisor('LG', 'teste')
tv1.sintoniza_canal('SBT')
tv1.sintoniza_canal('Globo')
tv1.aumenta_volume(90)
tv1.diminui_volume(110)
tv1.troca_canal('GlOBo')
tv1.troca_canal('AXN')
tv1
```

```python
import time

class Cronometro:
        
    def __init__(self, segundos = 0, minutos = 0, horas = 0):
        '''
        Cria uma representação do cronômetro

        Parâmetros:
        -----------
        segundos: int
            Valor referente aos segundos
        minutos: int
            Valor referente aos minutos
        horas: int
            Valor referente às horas
        '''
        self.segundos = segundos
        self.minutos = minutos
        self.horas = horas
        
    def __repr__(self):
        '''
        Formata a apresentação do cronômetro
        '''
        return "%.2d:%.2d:%.2d" %(self.horas, self.minutos, self.segundos)
    
    def incremento(self):
        '''
        Realiza a contagem (incremento) dos segundos/minutos/horas
        '''
        if self.segundos >= 60:
            self.segundos = 0
            self.minutos += 1
        if self.minutos >= 60:
            self.minutos = 0
            self.horas += 1
        self.segundos += 1
        
    def run(self):
        '''
        Inicia o cronômetro
        '''
        while True:  
            print(self)
            self.incremento()
            time.sleep(1)
            
#Exemplo de uso:
cronometro1 = Cronometro()
cronometro1
```

14. Faça uma classe ContaVip que difere da ContaCorrente por ter cheque especial (novo atributo) e é filha da classe ContaCorrente. Você precisa implementar os métodos para saque, transferência ou depósito?

```python
class Cliente:
    
    def __init__(self, nome, idade, email):
        '''
        Cria o cliente
        
        Parâmetros:
        -----------
        nome: str
            Nome do cliente
        idade: int
            Idade
        email: str
            Email
        '''
        self.nome = nome
        self.idade = idade
        self.email = email
        
    def __repr__(self):
        texto  = "Nome: {}\nIdade: {}\nE-mail: {}".format(
            self.nome,
            self.idade,
            self.email
        )
        return texto

    
class ContaCorrente:

    def __init__(self, cliente, saldo=0):
        '''
        Cria a representação da conta corrente

        Parametros:
        -----------
        cliente = object
            Objeto da classe Cliente
        '''
        self.cliente = cliente
        self.__saldo = saldo
    
    @property
    def saldo(self):
        return self.__saldo
        
    def __repr__(self):
        texto = self.cliente.__repr__() + "Saldo: {}".format(
            self.__saldo
        )
        
    def saque(self, valor):
        '''
        Realiza a dedução do valor na conta corrente, caso haja saldo suficiente
        
        Parametros:
        -----------
        valor = float
            Valor do saque
        '''
        if self.__saldo - valor >= 0:
            self.__saldo -= valor
        else:
            print('Saldo insuficiente.')
            
    def deposito(self, valor):
        '''
        Realiza a adição do valor na conta corrente
        
        Parametros:
        -----------
        valor = float
            Valor do depósito
        '''
        self.__saldo += valor
        
    def transferencia(self, other, valor):
        '''
        Realiza a transferência do valor da conta corrente de self para other
        
        Parametros:
        -----------
        other = object
            Objeto da classe Cliente
        valor = float
            Valor da transferência
        '''
        if self.__saldo - valor >= 0:
            self.__saldo -= valor
            other.__saldo += valor
        else:
            print('Saldo insuficiente.')
            
            
class ContaVip(ContaCorrente):
    
    def __init__(self, cliente, saldo=0, cheque_especial=0):
        '''
        Cria a representação da conta Vip contendo o atributo cheque_especial

        Parametros:
        -----------
        cliente = object
            Objeto da classe Cliente
        '''
        super().__init__(cliente, saldo)
        self.cheque_especial = cheque_especial
    
    def saque(self, valor):
        '''
        Realiza a dedução do valor na conta corrente, caso haja saldo suficiente
        
        Parametros:
        -----------
        valor = float
            Valor do saque
        '''
        if((self.saldo + self.cheque_especial) - valor >= 0):
            self.deposito(-valor)
        else:
            print('Saldo insuficiente.')
            
    def transferencia(self, other, valor):
        '''
        Realiza a transferência do valor da conta corrente de self para other
        
        Parametros:
        -----------
        other = object
            Objeto da classe Cliente
        valor = float
            Valor da transferência
        '''
        if((self.saldo+self.cheque_especial) - valor >= 0):
            self.deposito(-valor)
            other.deposito(valor)
        else:
            print('Saldo insuficiente.') 
            
#Exemplo de uso:
cliente1 = Cliente('Thiago', '123', 'thiago@teste')
cliente2 = Cliente('Lucas', '456', 'lucas@teste')
conta_thiago = ContaCorrente(cliente1, saldo=450)
conta_lucas = ContaVip(cliente2, saldo=1000, cheque_especial=500)
conta_thiago.transferencia(conta_lucas, 500)
conta_lucas.transferencia(conta_thiago, 1200)
print(conta_lucas.saldo)
conta_lucas.deposito(500)
print(conta_thiago.saldo)
```
